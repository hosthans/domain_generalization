# Demo of the PACS data loader

```python
import matplotlib.pyplot as plt
from torchvision import transforms as T
from pacs import get_data_loaders
```

```python
train_loader, test_loader = get_data_loaders(target_domain="cartoon", batch_size=16, transform=T.ToTensor())
```

```python
readable_labels = ["dog", "elephant", "giraffe", "guitar", "horse", "house", "person"]
```

```python
train_features, train_labels = next(iter(train_loader))
print(f"Feature batch shape: {train_features.size()}")
print(f"Labels batch shape: {train_labels.size()}")
img = train_features[0].squeeze()
label = train_labels[0]
plt.imshow(img.permute(1,2,0), cmap="gray")
plt.show()
print(f"Label: {label} {readable_labels[label]}")
```
